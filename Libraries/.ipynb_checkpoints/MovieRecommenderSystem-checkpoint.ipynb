```python
import pandas as pd
import numpy as np
from google.colab import files

r_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv("u.data", names=r_cols, sep='\t', usecols=range(3))
ratings.head()
```

```python
m_cols = ['movie_id', 'title']
movies = pd.read_csv("u.item", names=m_cols, usecols=range(2), sep='|', encoding="ISO-8859-1")
movies.head()
```

```python
ratings = pd.merge(movies, ratings)
ratings.head()
```

```python
user_ratings = ratings.pivot_table(index=['user_id'], columns=['title'], values='rating')
user_ratings.head()
```

```python
corr_matrix = user_ratings.corr()
corr_matrix.head()
```

```python
# wyeliminowano dziwne wartości
corr_matrix = user_ratings.corr(method='pearson', min_periods=100)
corr_matrix.head()
```

```python
my_ratings = user_ratings.loc[0].dropna() # loc jest tutaj jako user id
my_ratings.head()
```

```python
similiar_candidates = pd.Series()
for i in range(0, len(my_ratings.index)):
  print("Add simulars for: " + my_ratings.index[i])

  # uzyskaj podobne
  similiars = corr_matrix[my_ratings.index[i]].dropna()

  # skalowanie podobieństwa na podstawie jak dobrze oceniony został
  similiars = similiars.map(lambda x: x * my_ratings[i])

  # dodanie wyniku do listy
  similiar_candidates = similiar_candidates.append(similiars)

print("\nSORTING...")
similiar_candidates.sort_values(inplace=True, ascending=False)
print(similiar_candidates.head(20))
```

```python
# aby sie nie powtarzały filmy
similiar_candidates = similiar_candidates.groupby(similiar_candidates.index).sum()

similiar_candidates.sort_values(inplace=True, ascending=False)
similiar_candidates.head(10)
```

```python
# filter out movies I have already rated

filtered_similiars = similiar_candidates.drop(my_ratings.index)
filtered_similiars.head(10)
```
