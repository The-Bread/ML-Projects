### MatPlotLib

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-1, 6, .01)

# SCIPY.STATS.NORM.PDF:
# x: array_like quantities
# loc: location, parameter, default 0 - LOKACJA X
# scale: scale parameter(default=1)

plt.plot(x, norm.pdf(x))

# pdf - probability density function
# Funkcja gęstości prawdopodobieństwa
# nieujemna funkcja rzeczywista, określona dla rozkładu
# prawdopodobieństwa, taka że całka z tej funkcji, obliczona w
# odpowiednich granicach, jest równa prawdopodobieństwu
# wystąpienia danego zdarzenia losowego.

plt.show()
```

```python
# KILKA WYKRESÓW

plt.plot(x, norm.pdf(x)) # x must be array-like
plt.plot(x, norm.pdf(x, 1.0, 0.5))
plt.show()
```

```python
# ZAPISYWANIE DO PLIKU

plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0, 0.5))
plt.savefig('MojWykres.png', format='png')
```

```python
# MODYFIKACJA OSI X I/LUB Y

osie = plt.axes()
osie.set_xlim([0, 1]) # widoczny limit osi X
osie.set_ylim([0, 3]) # widoczny limit osi Y

osie.set_xticks([0.1, 0.2, 0.3, 0.4, 0.6, 0.8, 0.9, 1.1, 1.5])
osie.set_yticks([0, .3, .4, .5, .6, .7, .5, 3, 6])  # widoczne punkty na osiach

plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.5, 0.1))
plt.show()
```

```python
# DODANIE SIATKI - GRID

axes = plt.axes()
axes.set_xlim([0, 3])
axes.set_ylim([-1, 1])
axes.grid()
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0, .5))
plt.show()
```

```python
# ZABAWA Z KOLORAMI

# litera jako kolor, np: b - blue, r - red
# drugi znak jako kształt, np: o, -, +

# albo jako color='red', linestyle='dashed', marker='o' etc

plt.axes().grid()
plt.plot(x, norm.pdf(x), 'go')
plt.plot(x, norm.pdf(x, 1.0, .5), 'm+')
plt.show()
```

```python
# DODAJ LEGENDĘ

plt.axes().grid()
plt.xlabel("Opis osi X")
plt.ylabel('Opis osi Y')

plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 2.0, 1.2))

# LOC: lokacja legendy, do wyboru (0 - 10)
plt.legend(['Linia A', 'Linia B'], loc=10)
plt.show()
```

```python
# INNE STYLE

plt.xkcd()

figura = plt.figure()

data = np.ones(100) # tablica jedynek o rozmiarze
plt.plot(data)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
# 1 column - zmniejsza wysokosc wykresu
# 2 rows - zmniejsza szerokosc wykresu
# 3 index

ax.spines['right'].set_color('none') # usun kolory z osi
ax.spines['top'].set_color('none')
plt.xticks([]) # usun liczby z osi
plt.yticks([])
ax.set_ylim([-30, 10])

data = np.ones(100)
data[50:] -= np.arange(50) # zwraca jakas liczbe po jakims intervale

# od 70 ostatnich odejmij powoli 30
# ===PRZYPOMNIENIE===
# x = [1, 2, 3, 4, 5, 6]
# x[:3]  /// ([1, 2, 3])
# x[3:]  /// ([4, 5, 6])
# x[-2:] /// ([5, 6])

plt.annotate(
    'KIEDY SIĘ DOWIEDZIAŁAM, \nŻE ŁAKAMAKAFĄ TO ROCK \nTHE MICROPHONE',
    xy=(50, 0), arrowprops=dict(arrowstyle="->"), xytext=(0, -20)
)

plt.plot(data)

plt.xlabel('czas')
plt.ylabel('wiara w siebie')
```

```python
# PIE CHART

plt.rcdefaults()

wartosci = [12, 55, 13, 18, 2]
kolory = ['r', 'g', 'b', 'c', 'm']
wysuniecie = [0, 0, 0.3, 0.1, 0]
etykiety = ['Kasia', 'Asia', 'Basia', 'Ola', 'Ania']
plt.pie(wartosci, colors=kolory, labels=etykiety, explode=wysuniecie)
plt.title('Random Chart')
plt.show()
```

```python
# BAR CHART

wartosci = [12, 45, 40, 22, 11]
kolory = ['r', 'g', 'b', 'y', 'm']
plt.bar(range(0, 5), wartosci, color=kolory) # range musi sie zgadzać
plt.show()
```

```python
# WYKRES PUNKTOWY

from pylab import randn

X = randn(50)
Y = randn(50)
plt.scatter(X, Y)
plt.show()
```

```python
# HISTOGRAM

dochod = np.random.normal(27000, 10000, 5000)
plt.hist(dochod, 100)
plt.show()
```

```python
# BOX AND WHISKER PLOT

# używany do pokazania szerokości i krzywej (skew)

uniform_skewed = np.random.rand(100) * 100 - 40
wyzszy_outliner = np.random.rand(100) * 50 + 100
nizszy_outliner = np.random.rand(100) * -50 - 90
data = np.concatenate((uniform_skewed, wyzszy_outliner, nizszy_outliner))
plt.boxplot(data)
plt.show()
```
