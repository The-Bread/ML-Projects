```
import pandas as pd
import numpy as np
from google.colab import files

r_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv("u.data", names=r_cols, sep='\t', usecols=range(3))
ratings.head()
```

```
m_cols = ['movie_id', 'title']
movies = pd.read_csv("u.item", names=m_cols, usecols=range(2), sep='|', encoding="ISO-8859-1")
movies.head()
```

```
ratings = pd.merge(movies, ratings)
ratings.head()
```

```
user_ratings = ratings.pivot_table(index=['user_id'], columns=['title'], values='rating')
user_ratings.head()
```

```
corr_matrix = user_ratings.corr()
corr_matrix.head()
```
